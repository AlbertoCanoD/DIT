```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Import and show img1.

```python
from IPython.display import Image
Image("sample_data/img1.png")
```

Convert the image into grayscale.

```python
# Read the image
image = cv2.imread("sample_data/img1.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Check if not exist
if image is None:
    print("Image not found")
    exit(0)

colourplot = plt.imshow(image)
plt.axis('off')
plt.show()
```

Computing the histogram and show with the grayscaled image.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)

# Image
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Image')
ax1.axis("off")

# Histogram of coloured image
colours = ('b', 'g', 'r')
for i, col in enumerate(colours):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.xlabel("Colours levels")
    plt.ylabel("Relative frequency")
    ax2.set_title('RGB histogram')

plt.show()
```
