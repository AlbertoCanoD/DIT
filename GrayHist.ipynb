First we need to import basic packages.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Import and show img1.

```python
from IPython.display import Image
Image("img/img1.png")
```

Convert the image into grayscale.

```python
# Read the image and convert to grayscale
image = cv2.imread("img/img1.png", cv2.IMREAD_GRAYSCALE)

# Check if not exist
if image is None:
    print("Image not found")
    exit(0)

grayplot = plt.imshow(image, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

Computing the histogram and show with the grayscaled image.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)

# Image
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Imagen')
ax1.axis("off")

# Histogram of image
ax2.hist(image.ravel(), 256)
plt.xlabel("Niveles de gris")
plt.ylabel("Frecuencia relativa")
ax2.set_title('Histograma en escala de grises')

plt.show()
```
